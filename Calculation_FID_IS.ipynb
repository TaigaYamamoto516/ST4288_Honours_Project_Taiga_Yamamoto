```python
from pytorch_gan_metrics import ImageDataset
from torch.utils.data import DataLoader
from pytorch_gan_metrics import (get_inception_score, get_fid, get_inception_score_and_fid)
```

```python
path_to_dir = 'Path to images for FID Calculation'
dataset = ImageDataset(path_to_dir, exts=['png', 'jpg'])
loader = DataLoader(dataset, batch_size=50, num_workers=2)
use_torch = True
IS, IS_std = get_inception_score(loader, use_torch=True)
FID = get_fid(loader, 'Path to calculation data')
print('Inception score: ', IS)
print('Inception score std: ', IS_std)
print('Frechet Inception Distance: ', FID)
```
